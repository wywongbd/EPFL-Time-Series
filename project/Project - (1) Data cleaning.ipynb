### Read Data

```r
data <- read.csv(file="cleaned_data.csv", header=TRUE, sep=",")
```

```r
head(data)
```

```r
tail(data)
```

### Visualize Locations of bad quality data

MaxTemp with bad quality is recorded with 0 and -1. 

```r
plot(data$MaxTemp, type = 'l')
abline(0, 0)
```

```r
which(data$MaxTemp == 0)
bad_rows_index <- which(data$MaxTemp == 0)
```

```r
which(data$MaxTemp == -1)
bad_rows_index <- c(bad_rows_index, which(data$MaxTemp == -1))
```

```r
bad_rows_index
```

```r
bad_rows_index <- sort(bad_rows_index)
bad_rows_index
```

We can easily see that the largest gap is between 57086 and 4745.

```r
data[4745, ]
data[4746, ]
data[4747, ]
data[57085, ]
data[57086, ]
data[57087, ]
```

### Obtain subset of dataframe without bad quality rows

```r
final_cleaned_data <- data[4746:47085,]
```

```r
nrow(final_cleaned_data)
```

```r
str(final_cleaned_data)
```

### Obtain subset of dataframe WITH bad quality rows

```r
bad_rows_index
```

Basically, there are three main continuous segments in the bad data

```r
final_bad_data <- data[bad_rows_index,]
```

```r
nrow(final_bad_data)
```

```r
str(final_bad_data)
```

### Write data to new csv file

```r
write.table(final_cleaned_data, file = "final_cleaned_data.csv", quote = FALSE, sep = ",")
```
